```python
df = pd.read_csv("gRNA_in_locus.bed",sep="\t",header=None)
df['l'] = df[0]+":"+df[1].astype(str)+"-"+df[2].astype(str)
df['g'] = df[3]+":"+df[4].astype(str)+"-"+df[5].astype(str)
df.index = df[6]
df.head()
```

```python
abe = pd.read_csv("ABE_high_vs_low_mageck_RRA_results.sgrna_summary.txt",sep="\t")
abe['FDR'] = abe['FDR']
abe.index = [x.split("-")[-1] for x in abe.sgrna]
myFDR = abe['FDR'].to_dict()
df['FDR'] = df[6].map(myFDR)
df['FDR'] = df['FDR'].astype(str)
abe.head()
```

```python
locus = pd.DataFrame(df.groupby('l')['g'].agg(', '.join))
FDR = pd.DataFrame(df.groupby('l')['FDR'].agg(', '.join))
gRNA = pd.DataFrame(df.groupby('l')[6].agg(', '.join))
```

```python
locus_by_gRNA = pd.concat([locus,FDR,gRNA],axis=1)
```

```python
ABE_count = pd.read_csv("ABE_RRA_results.normalized.txt",sep="\t")
ABE_count.index = [x.split("-")[-1] for x in ABE_count.sgRNA]
ABE_count = ABE_count.drop(['sgRNA','Gene'],axis=1)
ABE_count.columns = [0,1,2,3]
ABE_count.head()
```

```python
# EmpiricalBrownsMethod(DataMatrix, Pvalues, extra_info = True)
def combine_p_value(x):
    """row apply for A_by_gRNA"""
    ## get count matrix
    if len(x[6].split(", "))==1:
        return float(x['FDR'])
    myCount = get_count_matrix(x[6].split(", ")).values
    ## get p-value vector
    p_value_vector = np.array([float(y) for y in x['FDR'].split(", ")])
    result = EmpiricalBrownsMethod(myCount, p_value_vector, extra_info = True)
    return result[0]
    
def get_count_matrix(gRNA_list):
    return ABE_count.loc[gRNA_list]
locus_by_gRNA['EBM_FDR']=locus_by_gRNA.apply(combine_p_value,axis=1)    
locus_by_gRNA.head()
```

```python
def lowest_FDR(x):
    return np.array([float(y) for y in x.split(", ")]).min()
locus_by_gRNA['min']=locus_by_gRNA.FDR.apply(lowest_FDR)
```

```python
locus_by_gRNA['EBM_FDR'] = locus_by_gRNA['EBM_FDR'].replace(0,1e-300)
locus_by_gRNA['log_EBM_FDR'] = locus_by_gRNA['EBM_FDR'].apply(lambda x:-np.log10(x))
locus_by_gRNA.to_csv("locus-level-FDR.csv")
```
